# Data visualization -->> Seaborn 

```python
# whatever we can do with matplot can also be done in seaplot 
# importing libraries 
import seaborn as sns 
import numpy as np
```

```python
# loading inbuilt dataset 
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# how sepal_lenth will grow with respect to sepal_width 
# with the help of visualization we can get data driven insights easily as compared statistics .. (without statistics)
sns.scatterplot(x=iris.sepal_length , y = iris.sepal_width)
```

```python
# whether there is a relationship between the sepal_lenth and the sepal_width 
# we can see if sepal_length is increasing then the sepal_width is also increasing 
```

```python
sns.scatterplot(x=iris.sepal_length , y = iris.petal_length)
# we can see if the sepal length of the flower is increasing the petal length will also be increase 
```

```python
# and aslo there is some set of the flowers where the sepal length increasing but the petal length is not increasing 
```

```python
# histograms
sns.displot(iris.sepal_length)
```

```python
sns.displot(iris['sepal_length']) # another way 
```

```python
# distplot also shows us the frquency or  no of counts as histogram 
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
# if we have to find that people who pays more bill will pay more tip ? how we can check 
sns.scatterplot(x='total_bill' , y = 'tip' , data = tips)
```

```python
# we can see if the bill is increasing than the tip size will also be increased 
# another method
sns.scatterplot(x= tips.total_bill , y = tips.tip)
```

```python
tips.head()
```

```python
# how many are smokers and non smokers
tips['smoker'].value_counts()
```

```python
# if we have to represent if person is smoker or not
sns.regplot(x='total_bill' , y = 'tip' , data = tips)
```

```python
# hue parameter is used to do separation 
sns.relplot(x='total_bill' , y = 'tip' , data = tips , hue ='smoker')
# hue is nothing but color hue will differntiate the data by using the color 
```

```python
sns.scatterplot(x='total_bill' , y = 'tip' , hue='smoker' , data = tips)
```

```python
# highest bill paid by a smoker and highest tip given by him 
```

```python
tips.head()
```

```python
# lets plot the size 
tips['size'].value_counts()
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data = tips , hue='size')
```

```python
# here color is making some sort of confusion for us how to deal with it using other parameters 
# lets use the style parameter which will try to make different shape 
sns.relplot(x='total_bill' , y = 'tip' , data = tips , style='size')
```

```python
# we want to find out if a person is coming for dinner or lunch 
sns.relplot(x='total_bill' , y = 'tip' , data = tips , hue='time' , style = 'size')
```

```python
# on which day more customer are coming 
# we have a categorical data so we will use catplot here 
# and on the x axis we must use the 'day' , because it is categorical data 
sns.catplot(x='day' , y = 'total_bill' , data = tips)
```

```python
# sat and sunday more people are coming 
```

```python
sns.catplot(x='smoker' , y = 'tip' , data = tips)
```

```python
sns.jointplot(x=tips.total_bill , y =tips.tip)
```

```python
# technically this is doing same as the scatter plot but on this total bill it is trying to make a histogrm
# jointplot is a combination of both histogram and scatter plot 
```

```python
# pairplot
sns.pairplot(tips)
```
