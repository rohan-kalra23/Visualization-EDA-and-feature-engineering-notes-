```python
# let's do some beautification
plt.scatter(x,y)
plt.xlabel('this is my x axis ') # used for x label 
plt.ylabel("this is my y axis") # used for y label 
plt.title('this is a scatter plot graph') # used for title
```

```python
plt.scatter(x,y , c='green')   # c is used to control color
plt.xlabel("this is x axis ")
plt.ylabel("this is y axis ")
plt.title("this is scatter plot")
plt.show()
```

```python
plt.scatter(x,y , c='salmon')   # c is used to control color
plt.xlabel("this is x axis ")
plt.ylabel("this is y axis ")
plt.title("this is scatter plot")
plt.show()
```

```python
# we can choose any color code by visit the site html color code and then use #color code to get the color in your chart 
plt.scatter(x,y , c= '#D133FF')
```

```python
# we can also control the intesity 
plt.scatter(x,y , c='r' , alpha = .5) # this is a faded one 
```

```python
plt.scatter(x,y,c='r' , alpha = .9) # this is darker so alpha used for the intensity of the color 
```

```python
# marker 
plt.scatter(x,y,c='r' , alpha = 1 , marker = 'x')
```

```python
plt.scatter(x,y,c='r' , alpha = 1 , marker = '*')
```

```python
# if we want to print all thing as a grid 
plt.scatter(x,y,c='r' , alpha = 1 , marker = "*")
plt.xlabel('x axis ')
plt.ylabel('y axis')
plt.title("this is scatter plot")
plt.grid() # used for the grid graph 
```

```python
plt.figure(figsize = (6,2)) # we can control the figure size 
plt.scatter(x,y,c='r' , alpha = 1 , marker = "*")
plt.xlabel('x axis ')
plt.ylabel('y axis')
plt.title("this is scatter plot")
plt.grid()
```

```python
# Barplot
# for scatter plot we need to have both the data as numerical data and for barplot one is numerical and other is categorical/string data  
```

```python
plt.bar(x,y)
```

```python
plt.figure(figsize=(6,3))
plt.bar(x,y , color = 'red' , alpha = .8 )
plt.xlabel('representing my categorical data')
plt.ylabel('representing my numerical data')
plt.title("this is my bar graph")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh(x,y , color = 'red' , alpha = .8 )
plt.xlabel('representing my numerical data')
plt.ylabel('representing my categorical data')
plt.title("this is my bar graph")
plt.grid()
plt.show()
# giving barh will make a horizontal graph whatever is y become x and whatever is x beocmes y 
```

```python
data = [1,2,3,4,5,6,5,5,6,3,7,8,7,8,2]
```

```python
# let's make a graph which on x axis it shows us data and on y axis it shows us the frequency
# histogram 
```

```python
plt.figure(figsize=(6,5))
plt.hist(data)
plt.show()
```

```python
# 1 has appeared 1 time 
# 2 , 2 time 
#3 , 2 time 
# 4 , 1 time and  6,7,8 (2-2 times )
```

```python
data
# unique dataset is on x axis and on y axis it shows us frequency 
```

```python
plt.figure(figsize=(7,4))
plt.hist(data)
plt.xlabel("representing unique value")
plt.ylabel("representing frequency")
plt.title("histogram example")
plt.grid()
plt.savefig("test.png") # it will save the image is the galary name as test
```

```python
# let's try to plot a 3D graph 
# to do that we have to createa a x , y , z coordinate
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
# 3D plot 
# x,y,z axis 
fig  = plt.figure()
ax=fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```

```python
# 3D plot 
# x,y,z axis 
fig  = plt.figure()
ax=fig.add_subplot(projection = '3d')
ax.scatter(x,y,z , color = 'red' , alpha = .8)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()    # z label function are not available 
```
